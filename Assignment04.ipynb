#### Creating a smaller data frame with the following 9 variables:

```python
df1 = df[['UNITID','INSTNM','CITY','STABBR','ZIP','ADM_RATE','UGDS','TUITIONFEE_IN','MN_EARN_WNE_P10']]
df1
```

Definition of the above 9 variables:

Variables

UNITID: Unit ID for institution, which is a unique number given to each postsecondary institutions surveyed by IPEDS.

LINSTNM: Institution name reported in IPEDS

CITY: Location of the Institute

STABBR: State postcode of the Institute

ZIP: ZIP code of the Institute

ADM_RATE: Accreditor for institution

UGDS: consists of the number of degree/certificate-seeking undergraduates enrolled in the fall as reported by the IPEDS

TUITIONFEE_IN: Fee structrue for Bachelor's degree in Personal And Culinary Services.

MN_EARN_WNE_P10: The average salary is calculated for all undergraduates at an institution receiving federal financial aid in a given academic year who are also working full-time but not in school.
